```python
# importing required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# reading the dataset and loading into a dataframe
df = pd.read_csv('../data/cyberbullying_tweets.csv')

df.head()
```

```python
# printing the shape: (rows, columns)
df.shape
```

```python
# distribution of datapoints into respective class labels
print(df['cyberbullying_type'].value_counts())

# plotting the class distribution after droping duplicates
plt.figure(dpi=110)
chart = sns.countplot(x="cyberbullying_type", data=df, palette="Set2")
axes = chart.axes
axes.set_title('Class Distribution')
axes.set_xlabel('Cyberbullying Type')
plt.show()
```

```python
# finding duplicate sentences
df[df['tweet_text'].duplicated()]
```

```python
# extracting indices of duplicate sentences
dup_indices = df[df['tweet_text'].duplicated()].index

# printing each set of duplicate sentences
for index in dup_indices:
    temp = df[df['tweet_text'] == df['tweet_text'].loc[index]]
    print(temp)
```

```python
# dropping the duplicate sentences
df = df.drop(index=dup_indices)

# resetting the index
df.reset_index()
    
# verifying if duplicate sentences are dropped
df[df['tweet_text'].duplicated()]
```

```python
df.shape
```

```python
# distribution of datapoints into respective class labels
print(df['cyberbullying_type'].value_counts())

# plotting the class distribution after droping duplicates
plt.figure(dpi=110)
chart = sns.countplot(x="cyberbullying_type", data=df, palette="Set2")
axes = chart.axes
axes.set_title('Class Distribution')
axes.set_xlabel('Cyberbullying Type')
plt.show()
```
