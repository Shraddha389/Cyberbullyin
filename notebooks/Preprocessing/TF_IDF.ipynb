```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv(r'..\..\data\train_data.csv')
df.sample(10)
```

```python
# vectorize tweet texts
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(df['clean_tweet'].values.astype('U'))
feature_names = vectorizer.get_feature_names()
df_tfidf = pd.DataFrame(vectors.toarray(), columns=feature_names)

# add original cyberbullying column to TF IDF vector data
df_tfidf['cyberbullying_type'] = df['cyberbullying_type']
```

```python
df_tfidf.sample(10)
```
