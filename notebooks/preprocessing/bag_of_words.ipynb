```python
# import required packages
import re
import pandas as pd
import numpy as np

# preprocess
from sklearn.feature_extraction.text import CountVectorizer

# local scripts
from text_utils import preprocess_corpus
```

```python
# load train set into dataframe
df = pd.read_csv('../../data/train_data.csv')

# printing the shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())
```

```python
# preprocess documents
# remove special characters, stopwords
# lemmatization
clean_tweets = preprocess_corpus(df.tweet_text)

clean_tweets.head()
```

```python
# instantiate a CountVectorizer
vectorizer = CountVectorizer()

# instantiate a CountVectorizer with min_df filter
# min_df: minimum document frequency
vectorizer_df_filter = CountVectorizer(min_df=30)

# train and construct bag of words
counts = vectorizer.fit_transform(clean_tweets)

# train and construct filtered bag of words
counts_df_filter = vectorizer_df_filter.fit_transform(clean_tweets)
```

```python
# extract bag of words as dataframe
df_bow = pd.DataFrame(counts.A, columns=vectorizer.get_feature_names_out())

# extract filtered bag of words as dataframe
df_bow_df_filter = pd.DataFrame(counts_df_filter.A, columns=vectorizer_df_filter.get_feature_names_out())

# include class label column
df_bow['cyberbullying_type'] = df['cyberbullying_type']
df_bow_df_filter['cyberbullying_type'] = df['cyberbullying_type']

# bag of words
# shape
display(df_bow.shape)

# first 5 datapoints
display(df_bow.head())

# filtered bag of words
# shape
display(df_bow_df_filter.shape)

# first 5 datapoints
display(df_bow_df_filter.head())
```
