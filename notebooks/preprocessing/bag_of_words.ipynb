```python
# import required packages
import re
import pandas as pd
import numpy as np

# preprocess
from sklearn.feature_extraction.text import CountVectorizer

# local scripts
from text_utils import preprocess_corpus
```

```python
# load train set into dataframe
df = pd.read_csv('../../data/train_data.csv')

# printing the shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())
```

```python
# preprocess documents
# convert null values to empty strings
df['clean_tweet'] = df['clean_tweet'].fillna('')

# # remove special characters, stopwords
# # lemmatization
# clean_tweets = preprocess_corpus(df.clean_tweet)

# clean_tweets.head()
```

```python
# instantiate a CountVectorizer
cv = CountVectorizer(min_df=10)

# train and construct bag of words
counts = cv.fit_transform(df['clean_tweet'].values)
```

```python
# extract bag of words as dataframe
df_bow = pd.DataFrame(counts.A, columns=cv.get_feature_names_out())

# include clickbait column
df_bow['cyberbullying_type'] = df['cyberbullying_type']

# shape
display(df_bow.shape)

# first 5 datapoints
df_bow.head()
```
